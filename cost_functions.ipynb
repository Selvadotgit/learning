## Huber loss

## huber loss = mse + map 

```python
threshold =1 #setting threshold above is higher error and below this threshold is lower error 
abs_error = np.abs(y_pred, y)
# mse
capped_error = np.minimum(abs_error, threshold)
mse = 0.5*capped_error**2

# map
mae = threshold*(abs_error - (0.5*threshold))

huber_loss = np.mean(mae +mse)
huber_loss
```

## binary cross entropy

```python
y = np.array([0,0,0,1,1,1])
prob = np.array([0.2,0.3,0.4,0.9,0,0.45])
```

```python
eps = 1e-15
prob_clipped = np.clip(prob, eps, 1 - eps)
cross_entropy = -np.sum(y*np.log2(prob_clipped) + (1-y)*np.log2(1-prob_clipped))
cross_entropy
```

## Hinge loss

```python
y= np.array([-1,1,-1,1])
decision_score = [100 , 0.5 , 0 , -5000]
hinge_loss = np.mean(np.maximum(0,1-(y*decision_score)))
hinge_loss
```

## Focal loss

```python
alpha = 0.2
gamma = 2
y = np.array([1,1,1 ,0])
true_prob= np.array([0.6,0.7,0.2,0.6])
pt = np.where(y==1 ,true_prob, 1-true_prob)
```

```python
pt
```

```python
focal_loss = np.mean(-alpha*((1-pt)**gamma)*np.log(pt))
```

## sillohutte score

```python
X = np.array([[1,2],[2,3],[3,4],[3,5]])
label = np.array([1,1,2,2])
def equalidian_dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))

def intradistance_btw_cluster(X,label,idx):
    cur_idx = np.where(label == label[idx])[0]
    cur_idx = cur_idx[cur_idx != idx]
    cur_clu = X[cur_idx]
    total = 0
    for each in cur_clu:
        total += equalidian_dist(X[idx],each)
    return total/len(cur_clu)
def nearest_cluster_distance(X, label , idx):
    labels_uni = np.unique(label)
    other_cluster = labels_uni[labels_uni != label[idx]]    
    min_dist = np.inf
    for clu in other_cluster:
        cur_clu_idx = np.where(label==clu)[0]
        cur_clu = X[cur_clu_idx]
        total=0
        for each in cur_clu:
            total+= equalidian_dist(each , X[idx])
        avg = total/len(cur_clu)
        min_dist = min(avg , min_dist)
    return min_dist
    
s =0
for idx in range(len(X)):
    a = intradistance_btw_cluster(X,label,idx)
    b  = nearest_cluster_distance(X,label, idx)
    s += (b-a)/max(b,a)
sillohutte_score = s/len(X)
sillohutte_score
        
```

## Davies -  bouldin index

```python
X = np.array([[1,2],[2,3],[3,4],[3,3]])
label = np.array([0,0,1,1])
centroid = {0:np.array([2,2]),1:np.array([3,3])}
def scatterred_dist(X,centroid , idx,label):
    cur_clu_idx = np.where(label==label[idx])[0]
    points = X[cur_clu_idx]
    total=0
    dbi = 0
    for p in points:
        total+=equalidian_dist(p,centroid)
    return total/len(points)
def davis_bouldin_index():
    cluster = np.unique(label)
    R = []
    for i in cluster:
        max_similarity = -np.inf
        s_i = scatterred_dist(X,centroid[i],i,label)
        for j in cluster:
            if i == j:
                continue
            s_j = scatterred_dist(X, centroid[j],j,label)
            mij = equalidian_dist(centroid[i],centroid[j])
            dbi  = (s_i+s_j)/mij
            max_similarity = max(max_similarity,dbi)
        R.append(max_similarity)
    return np.mean(R)
davis_bouldin_index()            
    
```
