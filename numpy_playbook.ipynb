## Matrix Decompositions

## Eigen values and eigen vector

```python
np.linalg.eig(a)
```

```python
eig_vals , eig_vec = np.linalg.eig(a)
```

```python
eig_vals, eig_vec
```

## singular values decompostion

```python
u , s , vh = np.linalg.svd(a)
```

```python
u
```

```python
s ## singular values you have to create diagonal elements with zero array called sigma
```

```python
sigma =  np.diag(s)
```

```python
vh
```

```python
(u @ sigma) @ vh
```

```python
u @ (sigma @ vh)
```

```python
vh @ (sigma @ u) # order is important
```
